```python
import pandas as pd

clonality_df = pd.read_csv("combined_clonality.csv")

clonality_df.head(10)
```

```python
line_counts_df = pd.read_csv("line_counts.csv")

line_counts_df["line_count"] = line_counts_df["line_count"] / 4
line_counts_df.head(10)
```

```python
merged_clonality_df=pd.merge(clonality_df, line_counts_df, on='SAMPLE')
merged_clonality_df.head(10)
```

# normalize

```python
merged_clonality_df["nIGH"] = (merged_clonality_df["nIGH"] / merged_clonality_df["line_count"]) * 1000000
merged_clonality_df["nIGK"] = (merged_clonality_df["nIGK"] / merged_clonality_df["line_count"]) * 1000000
merged_clonality_df["nIGL"] = (merged_clonality_df["nIGL"] / merged_clonality_df["line_count"]) * 1000000
merged_clonality_df["nTCRA"] = (merged_clonality_df["nTCRA"] / merged_clonality_df["line_count"]) * 1000000
merged_clonality_df["nTCRB"] = (merged_clonality_df["nTCRB"] / merged_clonality_df["line_count"]) * 1000000
merged_clonality_df["nTCRD"] = (merged_clonality_df["nTCRD"] / merged_clonality_df["line_count"]) * 1000000
merged_clonality_df["nTCRG"] = (merged_clonality_df["nTCRG"] / merged_clonality_df["line_count"]) * 1000000
merged_clonality_df["loadIGH"] = (merged_clonality_df["loadIGH"] / merged_clonality_df["line_count"]) * 1000000
merged_clonality_df["loadIGK"] = (merged_clonality_df["loadIGK"] / merged_clonality_df["line_count"]) * 1000000
merged_clonality_df["loadIGL"] = (merged_clonality_df["loadIGL"] / merged_clonality_df["line_count"]) * 1000000
merged_clonality_df["loadTCRA"] = (merged_clonality_df["loadTCRA"] / merged_clonality_df["line_count"]) * 1000000
merged_clonality_df["loadTCRB"] = (merged_clonality_df["loadTCRB"] / merged_clonality_df["line_count"]) * 1000000
merged_clonality_df["loadTCRD"] = (merged_clonality_df["loadTCRD"] / merged_clonality_df["line_count"]) * 1000000
merged_clonality_df["loadTCRG"] = (merged_clonality_df["loadTCRG"] / merged_clonality_df["line_count"]) * 1000000

merged_clonality_df.head(10)
```

```python
merged_clonality_df.to_csv('annie_normalized_clonality.csv',index=False)
```
