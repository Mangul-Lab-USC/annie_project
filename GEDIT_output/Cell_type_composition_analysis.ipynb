```python
clonality_df = pd.read_csv("../clonality/combined_clonality.csv")
line_counts_df = pd.read_csv("../clonality/line_counts.csv")
line_counts_df["line_count"] = line_counts_df["line_count"] / 4
normalized_clonality_df=pd.merge(clonality_df, line_counts_df, on='SAMPLE')
normalized_clonality_df["nIGH"] = (normalized_clonality_df["nIGH"] / normalized_clonality_df["line_count"]) * 1000000
normalized_clonality_df["nIGK"] = (normalized_clonality_df["nIGK"] / normalized_clonality_df["line_count"]) * 1000000
normalized_clonality_df["nIGL"] = (normalized_clonality_df["nIGL"] / normalized_clonality_df["line_count"]) * 1000000
normalized_clonality_df["nTCRA"] = (normalized_clonality_df["nTCRA"] / normalized_clonality_df["line_count"]) * 1000000
normalized_clonality_df["nTCRB"] = (normalized_clonality_df["nTCRB"] / normalized_clonality_df["line_count"]) * 1000000
normalized_clonality_df["nTCRD"] = (normalized_clonality_df["nTCRD"] / normalized_clonality_df["line_count"]) * 1000000
normalized_clonality_df["nTCRG"] = (normalized_clonality_df["nTCRG"] / normalized_clonality_df["line_count"]) * 1000000
normalized_clonality_df["loadIGH"] = (normalized_clonality_df["loadIGH"] / normalized_clonality_df["line_count"]) * 1000000
normalized_clonality_df["loadIGK"] = (normalized_clonality_df["loadIGK"] / normalized_clonality_df["line_count"]) * 1000000
normalized_clonality_df["loadIGL"] = (normalized_clonality_df["loadIGL"] / normalized_clonality_df["line_count"]) * 1000000
normalized_clonality_df["loadTCRA"] = (normalized_clonality_df["loadTCRA"] / normalized_clonality_df["line_count"]) * 1000000
normalized_clonality_df["loadTCRB"] = (normalized_clonality_df["loadTCRB"] / normalized_clonality_df["line_count"]) * 1000000
normalized_clonality_df["loadTCRD"] = (normalized_clonality_df["loadTCRD"] / normalized_clonality_df["line_count"]) * 1000000
normalized_clonality_df["loadTCRG"] = (normalized_clonality_df["loadTCRG"] / normalized_clonality_df["line_count"]) * 1000000
normalized_clonality_df.head(10)
```
